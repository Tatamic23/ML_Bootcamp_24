Linear Regression

```python
x =df[['reading_score', 'writing_score']]
y = df['math_score']

x_train, x_test, y_train, y_test = train_test_split(x , y, test_size=0.3, random_state=42)
```

```python
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
```

```python
y_pred = linear_model.predict(x_test)

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f'Mean Absolute Error:   {mae}')
print(f'Root Mean Squared Error:   {rmse}')
```

```python
plt.figure(figsize=(10,6))
plt.scatter(y_test,y_pred,alpha=0.7, color='pink')
plt.xlabel("Actual Math Score")
plt.ylabel('Predected Math Score')
plt.title('Actual vs Presicted Math Score')
plt.plot([y_test.min(),y_test.max()], [y_test.min(), y_test.max()], 'k--' , lw=2, color="purple")
plt.show()
```

Logictic Regression

```python
df['high_math_score'] = (df['math_score'] > df['math_score'].median()).astype(int)
x = df[['reading_score', 'writing_score']]
y = df['high_math_score']

x_train,x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
logical_model = LogisticRegression()
logical_model.fit(x_train,y_train)
```

```python
y_pred = logical_model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test,y_pred)

print(f'Accuracy: {accuracy}')
print('Confusion Matrix:')
print(conf_matrix)
print('Classification Report:')
print(class_report)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Pastel1', xticklabels=['Low', 'High'], yticklabels=['Low', 'High'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
